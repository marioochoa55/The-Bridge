### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np

class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

### Step 2. Create 3 differents Series, each of length 100, as follows: 
1. The first a random number from 1 to 4 
2. The second a random number from 1 to 3
3. The third a random number from 10,000 to 30,000

```python
sr_1 = pd.Series(np.random.randint(1, 5, size=100))
sr_2 = pd.Series(np.random.randint(1, 4, size=100))
sr_3 = pd.Series(np.random.randint(10000, 30001, size=100))

print(sr_1, sr_2, sr_3)
```

### Step 3. Let's create a DataFrame by joinning the Series by column

```python
df = pd.DataFrame({"col1": sr_1, "col2": sr_2, "col3": sr_3})
df
```
