### Step 13. What is the most frequent occupation?

```python
occupations = users["occupation"].value_counts()
print(occupations)
print("\n")
print(occupations.idxmax())
print("\n")
print(occupations.sort_values(ascending=False).iloc[[0]])
print("\n")
print(users["occupation"].mode())
```

```python
users["occupation"].describe()["top"]
```

### Step 18. What is the age with least occurrence?

```python
print(users["age"].value_counts().iloc[[-1]])
# print(neg.iloc[[0]] * -1)
users[["occupation", "age", "gender"]].groupby(["occupation", "age"]).agg('count')
```

```python
n_min = users["age"].value_counts().min()
print("Valor minimo:", n_min)
vc_users = users["age"].value_counts()
print(vc_users)
vc_users[vc_users == n_min]
```
