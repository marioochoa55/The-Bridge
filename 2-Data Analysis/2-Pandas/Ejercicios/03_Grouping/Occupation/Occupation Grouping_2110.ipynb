# Occupation

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
n_men = len(users[users.gender == 'M'])
n = len(users)
print(n_men / n * 100)
```

```python
users.groupby(['occupation'])['gender'].value_counts(normalize=True)
```

```python
def men_ratio(x):
    n_men = x[x.gender == 'M']['gender'].count()
    return n_men / len(x)

users.groupby('occupation').apply(men_ratio)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].aggregate(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender'])['age'].mean().unstack(level=1)
```

### Step 8.  For each occupation present the percentage of women and men

```python
users.groupby('occupation')['gender'].value_counts(normalize=True)
```

```python
for occ, group in users.groupby('occupation'):
    print("Esto son los {}".format(occ))
    print(group)
    print("\n\n")
```

```python
users.groupby('gender')['zip_code'].count() / len(users)
```

```python
users.groupby('occupation').apply(lambda x: x.groupby('gender').count()/len(x))[['age']]
```
