## Ejercicio 4: Flujos de Control (1.5 puntos) --> Calificación 1.5/1.5

1. Hacer un programa que permita saber si un año es bisiesto. Para que un año sea bisiesto debe ser divisible por 4 y no debe ser divisible por 100, excepto que también sea divisible por 400.
2. Hacer un programa que, dado un número, nos diga si es Real o Entero.
3. Declara una variable numérica que será una hora cualquiera del día. Implementa mediante sentencias `if/elif/else` la siguiente casuística:
   * Si es entre las 0 y las 8, print "Durmiendo"
   * Si es entre las 9 y las 18, print "Trabajando"
   * Si es entre las 19 y las 21, print "Clase"
   * Si es entre las 22 y las 24, print "Descanso"
   * En cualquier otro caso, print "Transporte o error"

```python
#APARTADO 1 
año = int(input("Introduce un año para comprobar si es bisiesto"))
if año%4 != 0:
    print("Este año no es bisiesto")
elif año%4 == 0 and año%100 != 0:
    print("Este año es bisiesto")
elif año%4 == 0 and año%100 == 0 and año%400 == 0:
    print("Este año es bisiesto")
elif año%4 == 0 and año%100 == 0 and año%400 != 0:
    print("Este año no es bisiesto")
```

```python
#APARTADO 2
num = input("Introduce un numero cualquiera")
if "." in num:
    print("Este numero es real")
else:
    print("Este número es entero")

```

```python
#Apartado 3 
hora = int(input("Introduce la hora"))
if (hora >= 1) and (hora <= 8):
    print("Durmiendo")
elif (hora >= 9) and (hora <= 18):
    print("Trabajando")
elif (hora >= 19) and (hora <= 21):
    print("Clase")
elif (hora >= 22) and (hora <= 24):
    print("Descanso")
else:
    print("Transporte o error")
```

## Ejercicio 5: Calculadora simple (2 Puntos) --> 2/2

Construye un programa que:

- Salude al usuario por su nombre
- Le pregunte si quiere sumar, restar, multiplicar o dividir
- Pida que introduzca 2 números de manera sucesiva
- Devuelva el resultado de manera formal con un mensaje apropiado 

**Ayuda**: No se puede dividir por 0

```python
nombre = input("Introduce tu nombre")
print("Hola", nombre)
operacion = input("Introduce la operación que quiere realizar: sumar, resta, multiplicacion o division")
numero1 = float(input("Introduce un numero cualquiera"))
numero2 = float(input("Introduce otro numero cualquiera"))
if operacion == "suma":
    print("El resultado de sumar", numero1, "y", numero2, "es igual a", numero1 + numero2)
elif operacion == "multiplicación":
    print("El resultado de multiplicar", numero1, "por", numero2, "es igual a", numero1 * numero2)
elif operacion == "resta":
    print("El resultado de restar", numero1, "menos", numero2, "es igual a", numero1 - numero2)
if operacion == "division" and numero2 != 0:
    print("El resultado de dividir", numero1, "entre", numero2, "es igual a", numero1 / numero2)

```
